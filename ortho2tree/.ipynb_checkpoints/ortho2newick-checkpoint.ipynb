```python
import ete3
import xml.etree.cElementTree as ET
```

```python

tree = ET.parse('orthoxmlexample.txt')

root = tree.getroot()

print(root.tag)
gene2taxid ={}
gene2species={}
for species in root.getchildren():
    if 'species' in species.tag:
        name = species.items()[0][1]
        taxid = species.items()[1][1]
        print(name+taxid)
        for db in species.getchildren():
            print(db)
            for genes in db.getchildren():
                print(genes.tag)
                if 'genes' in genes.tag:
                    for gene in genes.getchildren():
                        if 'gene' in gene.tag:
                            gene2species[gene.get('id')]= name
                            gene2taxid[gene.get('id')]= taxid
```
